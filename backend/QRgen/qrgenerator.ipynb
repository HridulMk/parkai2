```python
import qrcode
import os

def generate_qr_code(data, file_name='my_qr.png'):
    # Create QR code instance
    qr = qrcode.QRCode(
        version=1, # 1 to 40 - controls the size of the QR Code
        error_correction=qrcode.constants.ERROR_CORRECT_H, # Higher error correction
        box_size=10, # Size of each box in pixels
        border=4, # Thickness of the border
    )
    
    # Add data to the QR code
    qr.add_data(data)
    qr.make(fit=True)
    
    # Create an image from the QR Code instance
    img = qr.make_image(fill_color="black", back_color="white")
    
    # Save the generated QR code image
    img.save(file_name)
    print(f"QR Code generated and saved as {file_name}")
    
    return img
```

```python
# Example usage:
data_to_encode = "https://www.example.com"
image_name = "example_qr_code.png"

img = generate_qr_code(data_to_encode, image_name)
img # Display image in the notebook
```
